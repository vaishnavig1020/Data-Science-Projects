```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
# Loading dataset 
fifa = pd.read_csv('players_20.csv')
```

```python
# Looking 1st five dataset
fifa.head()
```

```python
# looking names of the all columns present in the dataset
for col in fifa.columns:
    print(col)
```

```python
# looking no. of rows and columns
fifa.shape
```

```python
# looking no. of the frequency of nationality of the players
fifa['nationality'].value_counts()
```

```python
# looking top 10 largest nationality of the players
fifa['nationality'].value_counts()[0:10]
```

```python
# looking top 5 largest nationality of the players
fifa['nationality'].value_counts()[0:5]
```

```python
# list the names of the top 5 largerst nationality of the players
# looking top 10 largest nationality of the players
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# making a bar-plot of the top 5 nationality
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='r')
plt.show()
```

```python
# Making a Pie-chart top 5 nationality
plt.figure(figsize=(8,8))
plt.pie(fifa['nationality'].value_counts()[0:5],labels=fifa['nationality'].value_counts().keys()[0:5],autopct='%0.1f%%')
plt.show()
```

```python
# Players with their names and salary
player_salary=fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
# Arranging the player_salary with decending
player_salary.sort_values(by = ['wage_eur'],ascending=False)[0:10]
```

```python
# Making a Bar graph for the 1st five players w.r.t their salary
plt.figure(figsize=(8,8))
plt.bar(list(player_salary['short_name'][0:5]),list(player_salary['wage_eur'][0:5]),color=['green','red','blue','pink','yellow'])
plt.show()
```

```python
# Making a pie-chart for the 1st five players w.r.t their salary
plt.figure(figsize=(6,6))
plt.pie(player_salary['wage_eur'][0:5],labels=player_salary['short_name'][0:5],autopct='%0.1f%%')
plt.show()
```
