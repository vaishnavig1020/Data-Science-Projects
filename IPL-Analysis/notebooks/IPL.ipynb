```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
# Loading a ipl matches dataset
ipl=pd.read_csv('matches.csv')
```

```python
# Having a glance at 1st five records of dataset
ipl.head()
```

```python
# looking for the no. of the rows and columns of the dataset
ipl.shape
```

```python
# Getting the freqency of the most man of the match award
ipl['Player_of_Match'].value_counts()
```

```python
#Gettind the top 10 players with most man of the match award
ipl['Player_of_Match'].value_counts()[0:10]
```

```python
# Getting top 5 players with most man of the match
ipl['Player_of_Match'].value_counts()[0:5]
```

```python
# List of the top 5 players name with man of the match award
list(ipl['Player_of_Match'].value_counts()[0:5].keys())
```

```python
# Bar-plot for the most top 5 players of the man of the match
plt.figure(figsize=(5,5))
plt.bar(list(ipl['Player_of_Match'].value_counts()[0:5].keys()),list(ipl['Player_of_Match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
# Getting Frequency of winner column
ipl['Winner'].value_counts()
```

```python
# Find out the toss win w.r.t team
ipl['Toss_Winner'].value_counts()
```

```python
# Extracting the record wherre the team won the match by batting first
batting_first = ipl[ipl['Win_by_Runs']!=0]   # New dataset
```

```python
# Looking 1st five teams won by the batting first
batting_first.head()
```

```python
# Making a histogram of the team won by the taking batting first
plt.figure(figsize=(4,4))
plt.hist(batting_first['Win_by_Runs'],bins=30)
plt.title('Winners of the Match')
plt.xlabel("Runs")
plt.ylabel('No. of Match')
plt.show()
```

```python
# Finding out the no. of win w.r.t. teams won the match after taking batting first
batting_first['Winner'].value_counts()
```

```python
# Makiing a bar-plot for the top 3 teams wo the match by taking batting first
plt.figure(figsize=(10,5))
plt.bar(list(batting_first['Winner'].value_counts()[0:3].keys()),list(batting_first['Winner'].value_counts()[0:3]),color=['blue','red','yellow'])
plt.show()
```

```python
# Making a Pie-Chart for the teams won by the taking batting first.
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['Winner'].value_counts()),labels=list(batting_first['Winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
# Extracting the records of the match won by second batting
batting_second= ipl[ipl['Win_by_Wickets']!=0]
```

```python
# looking for the 1st five team
batting_second.head()
```

```python
# Making a histogram no. of the frequency of won w.r.t no. of wickets 
plt.figure(figsize=(5,5))
plt.hist(batting_second['Win_by_Wickets'],color='red',bins=30)
plt.show()
```

```python
# finding out the frequency of the no. wins after batting
batting_second["Winner"].value_counts()
```

```python
# plotting a bar plot of top 3 teams which won the match after batting
plt.figure(figsize=(5,5))
plt.bar(list(batting_second["Win_by_Wickets"].value_counts()[0:3].keys()),list(batting_second["Winner"].value_counts()[0:3]))
plt.title("Winner of the Match")
plt.xlabel("No. of match")
plt.ylabel("Won by wickets")
plt.show()
```

```python
# Making a pie-chat
plt.figure(figsize=(5,5))
plt.pie(list(batting_second["Winner"].value_counts()),labels=list(batting_second["Winner"].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
